```python
print("==========================================")
print("           MECABOT INTELIGENTE            ")
print("==========================================")
historial = [] # Lista que empieza vacía
continuar = "si"
while continuar == "si":
    print("\n--- MENÚ MECÁNICO ---")
    print("1. Revisar Temperatura")
    print("2. Ver Historial de Fallas")
    print("3. Salir")
    opcion = input("Selecciona (1, 2 o 3): ")
    if opcion == "1":
        temp = float(input("¿Temperatura del motor en °C?: "))
        if temp > 105:
            print(" IA Dice: ¡Alerta de sobrecalentamiento!")
            historial = historial + ["Motor Caliente"] # Agrega a la lista
        elif temp < 70:
            print(" IA Dice: Motor muy frío.")
        else:
            print(" IA Dice: Temperatura normal.")
    elif opcion == "2":
        if historial == []:
            print(" IA Dice: No hay fallas registradas. ¡Auto sano!")
        else:
            print("Fallas detectadas:", historial)
    elif opcion == "3":
        print(" IA Dice: Apagando MecaBot...")
        continuar = "no"
    else:
        print("Opción incorrecta.")
    if opcion != "3":
        continuar = input("\n¿Hacer otra revisión? (si/no): ")
print("==========================================")
print("Programa realizado por Meredith Aguirre NL = 2")
```
