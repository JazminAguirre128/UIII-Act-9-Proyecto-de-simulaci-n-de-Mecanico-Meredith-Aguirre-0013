```python
print("==========================================")
print("            MECABOT INTELIGENTE           ")
print("==========================================")

# Estructuras de datos que inician vacías
historial_diagnosticos = []
reporte_actual = {}

while True:
    print("\n--- MENÚ MECABOT ---")
    print("1. Realizar Diagnóstico Inteligente")
    print("2. Ver Historial de Autos Revisados")
    print("3. Salir")

    opcion = input("Selecciona una opción (1-3): ")

    if opcion == "1":
        print("\n--- NUEVOS DATOS DEL VEHÍCULO ---")
        marca = input("Marca del auto: ")
        temp = float(input("¿Temperatura del motor en °C?: "))
        ruido = input("¿El motor hace un ruido extraño? (si/no): ")

        # La "IA" toma decisiones basadas en las variables
        if temp > 105 and ruido == "si":
            diagnostico = "¡Alerta Crítica! Motor sobrecalentado y posible daño interno."
        elif temp > 105:
            diagnostico = "Falla: Sobrecalentamiento. Revisar líquido refrigerante."
        elif temp < 70:
            diagnostico = "Advertencia: Motor muy frío. Termostato pegado abierto."
        elif ruido == "si":
            diagnostico = "Advertencia: Posible falla en las bielas o bandas."
        else:
            diagnostico = "Sistema Óptimo. El auto está en perfectas condiciones."

        print(f"\n🤖 IA MecaBot Dice: {diagnostico}")

        # Creamos el diccionario del auto actual
        reporte_actual = {
            "auto": marca,
            "temperatura": temp,
            "resultado": diagnostico
        }
        # Agregamos el diccionario a la lista de historial
        historial_diagnosticos.append(reporte_actual)

    elif opcion == "2":
        print("\n--- HISTORIAL DE REVISIONES ---")
        if len(historial_diagnosticos) == 0:
            print("🤖 IA Dice: El historial está vacío. ¡No hay autos registrados aún!")
        else:
            # Ciclo FOR para recorrer la lista de diccionarios
            contador = 1
            for registro in historial_diagnosticos:
                print(f"\n🚗 Auto #{contador}")
                print(f"  • Marca: {registro['auto']}")
                print(f"  • Temp: {registro['temperatura']}°C")
                print(f"  • Dictamen de IA: {registro['resultado']}")
                contador = contador + 1

    elif opcion == "3":
        print("\n🤖 IA Dice: Apagando MecaBot... ¡Buen viaje!")
        break # Rompe el ciclo while True para salir

    else:
        print("❌ Opción incorrecta. Intenta de nuevo.")

print("\n==========================================")
print("Programa realizado por Meredith Aguirre NL = 2")
print("==========================================")
```
